Трансформеры —  тип нейросетей, направленный на решение последовательностей с легкой обработкой дальнодействующих зависимостей. Используют механим внимания, что позволяет улучшить эффективность. Применяется в области обработки естественной речи (NLP) для решения задач машинного перевода, написания текстов, генерации компьютерного кода.

В задачах они сходны с рекуррентными нейронными сетями (RNN). RNN обрабатывают слова последовательно, подвержены проблеме исчезающего градиента. Трансформеры не обрабатывают последовательности по порядку. Например, если исходные данные — текст, то им не нужно обрабатывать конец предложения после обработки начала. Благодаря этому такую нейросеть можно распараллелить и обучить значительно быстрее.

Эта архитектура позволяет избавиться от последовательных вычислений, т.е. можно производить вычисления параллельно, а не ждать окончания предыдущего шага.

* Позиционные энкодеры

В трансформерах используются позиционные энкодеры. Все слова исходной последовательности и добавить номер (или синусоидальную функцию) в его позиционном расположении. Тем самым сохраняется порядок слов в виде данных, а не в виде структуры. Это помогает распараллеливанию процесса обучения.

Трансформер состоит из двух частей: энкодера и декодера. Энкодер кодирует входную последовательность (например, слов) в вектор (т.е. набор чисел), декодер — декодирует ее в виде новой последовательности.

Энкодер и декодер в стандартном трансформере содержат по шесть слоев, то есть наборов математических операций. Это число подобрано опытным путём.

* Самовнимание

Внутри каждого слоя энкодера есть две составные части: слой внимания и нейросеть с прямой связью.

Энкодер зашифрует все предложение в единый вектор-контекст. Вектор-контекст — сумма векторов отдельных слов. Перед тем, как складывать слова, вектора важных слов умножим на большие числа. Таким образом мы искусственно повысим значимость важных слов на последующих шагах работы алгоритма. Это умножение — и есть механизм внимания.

В трансформере тоже работает подобное, но не в кодирующей части. Здесь применяется self-attention («внимание на себя»). Внимание энкодера обращено на предыдущий слой энкодера, то есть как бы на себя же, но в прошлом.

На самом первом слое предыдущих нет, поэтому трансформер получает «сырые» вектора слов и концентрирует внимание на некоторых из них. Новый вариант входного предложения, где какие-то слова уже помечены как важные, попадает в нейросеть с прямой связью, зачем она там — в следующем разделе. Оттуда переделанный вектор попадает на следующий слой энкодера, и так повторяется 6 раз.

В каждом слое энкодера после слоя внимания идет слой «с прямой связью».

Он нужен, для моделирования более сложных функции. Функции на слое внимания энкодера — линейные. Эта проблема решается, если между линейными слоями внимания добавить слои с нелинейными функциями активации.

Самовнимание помогает нейронным сетям устранять неоднозначность слов.
